### Script for data exploration

```python
from pathlib import Path
import csv
from PIL import Image
from random import randint

data_path = Path('../data')
train_data_path = data_path / "train_data.csv"
test_data_path = data_path / "test_data.csv"

def image_extractor(dataset, img_id=None):
    """Function to extract and display a image from dataset"""
    with open(dataset) as file:
        images = file.readlines() 
        indx = randint(1, len(images)) if img_id == None else img_id # get random index if none provided
        pixels = list(map(int, images[indx].split(','))) # get the pixel values and label from dataset
        label, pixels = pixels[0], pixels[1:] # convert the label to character and pixel values to list

        img = Image.new("L", (28, 28)) # create a new image object with the given size and mode (L for grayscale) 
        img.putdata(pixels) # put pixel values into the image object
        print(f"ID: {indx}, Label: {chr(label + 64)}") 
        img.show()
    
    
```

```python
image_extractor(test_data_path)
```
