### Script for calculation model learning time

```python
def prod(lst: list) -> int:
    prod = 1
    for element in lst: prod *= element
    return prod

image_size = 28
layer1_size = 32
layer2_size = 16
layers = [layer1_size, layer2_size]
samples = 140 * 10**3
epohes = 30

speed = 10 * 10**6

time = (image_size ** 2 * prod(layers) * epohes * samples + prod(layers)) / speed / 3600
time
```
